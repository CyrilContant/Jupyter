### Tracer le graphe représentatif d'une fonction en python

Pour tracer le graphe représentatif d'une fonction en python, nous aurons besoin de deux bibliothèques : **NumPy** et **matplotlib.pyplot**

```python
import numpy as np
import matplotlib.pyplot as plt
```

Le principe est assez simple : nous allons en fait représenter un nuage de points, suffisamment proches les uns des autres pour donner l'impression d'une continuité parfaite. Ainsi, il nous faut définir :

- une liste ou un tableau des abcisses

- une liste ou un tableau des ordonnées, de même longueur que celle des abcisses.

```python
X=np.linspace(0,20,1000)
Y=np.cos(X)
```

Nous avons utilisé ici la fonction linspace, qui prend habituellement trois argument : une valeur de départ $x_i$, une valeur de fin $x_f$, et un nombre de points $N$. Cette fonction renvoie simplement un tableau contenant $N$ valeurs, régulièrement réparties entre $x_i$ et $x_f$.

Les tableaux présentent un intérêt non négligeable en modélisation : nous pouvons leur appliquer directement les fonctions qui, normalement, s'appliquent aux élements qu'ils contiennent. Par exemple, ici, nous avons directement appliqué la fonction $cos$ au tableau $X$, et cela nous permet de créer un tableau $Y$ contenant le cosinus de l'ensemble des valeurs du tableau $X$. Cela nous épargne le parcours de la liste $X$ via une boucle.

Nous pouvons maintenant tracer le graphe :

```python
plt.plot(X,Y)
plt.show()
```
